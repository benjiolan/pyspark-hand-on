### Broadcast:

```python
impuesto = sc.broadcast(10)
```

```python
ventas_sin_impuesto = importar.map(lambda x: float(x) - impuesto.value)
```

```python
ventas_sin_impuesto.take(5)
```

```python
impuesto.destroy()
```
