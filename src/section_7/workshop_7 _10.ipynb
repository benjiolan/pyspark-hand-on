```python
%load_ext autoreload
%autoreload 2

import os
path = os.environ.get("HOME")
os.chdir(f"{path}/github/pyspark-hand-on")
from src.utils.utils import spark_session
from pyspark.sql.functions import col
spark = spark_session()
```

```python
df.dropna(subset=["views"]).count()
```

```python
df.dropna(subset=["views", "dislikes", ]).count()
```

```python
from pyspark.sql.functions import col
df.orderBy(col("views")).select(col("views"), col("likes"), col("dislikes")).show()
```

```python
df.fillna(0).orderBy(col("views")).select(col("views"), col("likes"), col("dislikes")).show()
```

```python
df.fillna(0, subset=["likes", "dislikes"]).orderBy(col("views")).select(col("views"), col("likes"), col("dislikes")).show()
```
