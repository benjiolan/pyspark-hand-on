```python
%load_ext autoreload
%autoreload 2

import os
path = os.environ.get("HOME")
os.chdir(f"{path}/github/pyspark-hand-on")
from src.utils.utils import spark_session
from pyspark.sql.functions import col
spark = spark_session()
```

```python
df = spark.read.parquet("./data/datos.parquet")
df.printSchema()
```

```python
## Define el nuemro de particiones del DF
df1 = df.repartition(2)
```

```python
## Guardar particiones en formato csv
df1.write.format("csv").option("sep", "|").mode("append").save("./output/data_csv")
```

```python
## Cambiar el numero de particiones
df1.coalesce(1).write.format("csv").option("sep", "|").save("./output/data_csv1")
```
