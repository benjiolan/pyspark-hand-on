#### Expr col:

```python
df.select(col("likes"), col("dislikes"), (col("likes")-col("dislikes")).alias("aceptacion")).show()
```

#### Select Expr:

```python
df.selectExpr("likes", "dislikes", "(likes - dislikes) as aceptacion").show()
```
