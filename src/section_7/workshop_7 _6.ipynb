### dropDuplicates

```python
new_df = spark.createDataFrame([(1, "azul", 567), (2, "rojo", 487), (1, "azul", 345), (2, "verde", 783)]).toDF("id", "color", "importe")
```

```python
new_df.show()
```

```python
new_df.dropDuplicates(["id", "color"]).show()
```
