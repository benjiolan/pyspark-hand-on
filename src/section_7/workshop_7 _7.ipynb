# Transformaciones - funciones sort & orderBy

```python
%load_ext autoreload
%autoreload 2

import os
path = os.environ.get("HOME")
os.chdir(f"{path}/github/pyspark-hand-on")
from src.utils.utils import spark_session
from pyspark.sql.functions import col
```

```python
spark = spark_session()
```

```python
df = spark.read.parquet("./data/datos.parquet")
df.printSchema()
```

```python
df = df.select(col("likes"), col("views"), col("video_id"), col("dislikes")).dropDuplicates(["video_id"])
df.show()
```

```python
## Sort
df.sort("likes").show()
```

```python
from pyspark.sql.functions import desc
```

```python
## Sort desc
df.sort(desc("likes")).show()
```

```python
##orderBy
df.orderBy(desc("views")).show()
```

```python
df.orderBy(col("likes").desc()).show()
```

```python
dataframe = spark.createDataFrame([(1, "azul", 567), (2, "rojo", 487), (1, "azul", 345), (2, "azul", 783)]).toDF("id", "color", "importe")
```

```python
dataframe.show()
```

```python
dataframe.orderBy(col("color").desc(), col("importe")).show()
```

```python
## funcion limit
top_10 = df.orderBy(col("views").desc()).limit(10)
top_10.show()
```
