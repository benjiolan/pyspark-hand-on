# Transformaciones - withColumn & withColumnRenamed

```python
%load_ext autoreload
%autoreload 2

import os
path = os.environ.get("HOME")
os.chdir(f"{path}/github/pyspark-hand-on")
from src.utils.utils import spark_session
from pyspark.sql.functions import col
```

```python
spark = spark_session()
```

```python
df = spark.read.parquet("./data/datos.parquet")
df.printSchema()
```

```python
## withColumn
df_valoracion = df.withColumn("valoracion", col("likes")-col("dislikes"))
```

```python
df_valoracion.printSchema()
```

```python
df_valoracion1 = df_valoracion.withColumn("valoracion", col("likes")-col("dislikes")).withColumn("res_div", col("valoracion")%10)
df_valoracion1.printSchema()
```

```python
df_valoracion1.select(col("likes"), col("dislikes"), col("valoracion"), col("res_div")).show()
```

```python
#withColumnRenamed
df_renamed = df.withColumnRenamed("video_id", "id")
```

```python
df_renamed.printSchema()
```

```python
df_error = df.withColumnRenamed("not_exists", "other_name")
```
