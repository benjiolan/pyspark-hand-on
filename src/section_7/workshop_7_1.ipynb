Un dataframe a partir de un rdd con schema:

```python
rdd1 = sc.parallelize([(1, "Jose",35.5), (2, "Teresa", 54.3), (3, "Katia", 12.7)])
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType
```

```python
## Via 1:
schema = StructType(
    [
        StructField("id", IntegerType(), True),
        StructField("nombre", StringType(), True),
        StructField("saldo", DoubleType(), True)
    ]
)
```

```python
## Via 2:
schema2 = "`id` INT, `nombre` STRING, `saldo` DOUBLE"
```

```python
df1 = spark.createDataFrame(rdd1, schema=schema)
```

```python
df1.printSchema()
```

```python
df1.show()
```

```python
df2 = spark.createDataFrame(rdd1, schema=schema2)
```

```python
df2.printSchema()
```

```python
df2.show()
```
