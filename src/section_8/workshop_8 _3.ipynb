```python
%load_ext autoreload
%autoreload 2

import os
path = os.environ.get("HOME")
os.chdir(f"{path}/github/pyspark-hand-on")
from src.utils.utils import spark_session
spark = spark_session()
```

```python
df = spark.read.parquet("./data/estudiantes.parquet")
df.printSchema()
```

```python
df.show()
```

```python
from pyspark.sql.functions import min, max, col, avg, col
```

```python
df.groupBy("graduacion").pivot("sexo").agg(avg("peso")).show()
```

```python
df.groupBy("graduacion").pivot("sexo").agg(avg("peso"), min("peso"), max("peso")).show()
```

```python
df.groupBy("graduacion").pivot("sexo", ["M"]).agg(avg("peso"), min("peso"), max("peso")).show()
```
