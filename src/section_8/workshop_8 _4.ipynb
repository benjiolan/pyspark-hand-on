## 69. Agregación con pivote:

```python
%load_ext autoreload
%autoreload 2

import os
path = os.environ.get("HOME")
os.chdir(f"{path}/github/pyspark-hand-on")
from src.utils.utils import spark_session
spark = spark_session()
```

```python
workers = spark.read.parquet("./data/empleados.parquet")
workers.show()
```

```python
departments = spark.read.parquet("./data/departamentos.parquet")
departments.show()
```

```python
from pyspark.sql.functions import col
```

## Inner join:

```python
## By default join uses inner (Implicit)
inner_df = workers.join(departments, col("num_dpto")==col("id"))
inner_df.show()
```

```python
## Explicit
joined_df = workers.join(departments, col("num_dpto")==col("id"), "inner")
joined_df.show()
```

```python
## Explicit other way
joined_df = workers.join(departments).where(col("num_dpto")==col("id"))
joined_df.show()
```

## Left outer join:

```python
# leftouter
workers.join(departments, col("num_dpto")==col("id"), "leftouter").show()
```

```python
# left_outer
workers.join(departments, col("num_dpto")==col("id"), "left_outer").show()
```

```python
## left
workers.join(departments, col("num_dpto")==col("id"), "left").show()
```

## Rigth outer join:

```python
# rightouter
workers.join(departments, col("num_dpto")==col("id"), "rightouter").show()
```

```python
# right_outer
workers.join(departments, col("num_dpto")==col("id"), "right_outer").show()
```

```python
# right
workers.join(departments, col("num_dpto")==col("id"), "right").show()
```

## Full outer join:

```python
workers.join(departments, col("num_dpto")==col("id"), "outer").show()
```

## Left anti join:

```python
workers.join(departments, col("num_dpto")==col("id"), "left_anti").show()
```

```python
## No Right_anti available
departments.join(workers, col("num_dpto")==col("id"), "left_anti").show()
```

## Left semi join:

```python
workers.join(departments, col("num_dpto")==col("id"), "left_semi").show()
```

## Cross join:

```python
cross_join = workers.crossJoin(departments)
```
