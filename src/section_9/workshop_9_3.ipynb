## 86. Funciones when, coalesce, lit:

```python
%load_ext autoreload
%autoreload 2

import os
path = os.environ.get("HOME")
os.chdir(f"{path}/github/pyspark-hand-on")
from src.utils.utils import spark_session

spark = spark_session()
```

```python
df = spark.read.parquet("./data/nombres")
df.printSchema()
```

```python
df.show(truncate=False)
```

```python
from pyspark.sql.functions import col, when, lit, coalesce
```

```python
## when
df.select(
    col("nombre"),
    when(col("pago")==1, "pagado").when(col("pago")==2, "sin pagar").otherwise("sin iniciar").alias("pago")
).show()
```

#### coalesce & lit:

```python
df.select(
    coalesce(col("nombre"), lit("sin nombre")).alias("nombre")
).show()
```
